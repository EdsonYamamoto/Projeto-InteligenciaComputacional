```python
# importar as bibliotecas do numpy, seaborn, pandas
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import re
```

```python
dfCerveja = pd.read_csv('Consumo_cerveja.csv', index_col=0)
display(dfCerveja.head())
```

```python
#regex para trocar ',' por '.'
regex_pat = re.compile(r',', flags=re.IGNORECASE)
```

```python
#Trocar , por pontos e enteder que é um número
dfCerveja['Temperatura Media (C)'] = pd.to_numeric(dfCerveja['Temperatura Media (C)'].replace({regex_pat: '.'}))
dfCerveja['Temperatura Minima (C)'] = pd.to_numeric(dfCerveja['Temperatura Minima (C)'].replace({regex_pat: '.'}))
dfCerveja['Temperatura Maxima (C)'] = pd.to_numeric(dfCerveja['Temperatura Maxima (C)'].replace({regex_pat: '.'}))
dfCerveja['Precipitacao (mm)'] = pd.to_numeric(dfCerveja['Precipitacao (mm)'].replace({regex_pat: '.'}))
```
